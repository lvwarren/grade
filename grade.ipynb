```python
def gen_grades(μ, σ, size, missed, policy):

    grades = norm.rvs(μ, σ, size=size)
    
    if policy.lower() == 'orthodox':
        for g in range(missed):
            grades[g]=0
    elif policy.lower() == 'novel':
        for g in range(missed):
            grades[g]=max_grade/2
    else:
        print('unknown policy')

    mu, std = norm.fit(grades)
    
    # Generate a equally spaced set of x values from 0 to max_grade
    x = np.linspace(0, max_grade, sample_size)

    # Generate the probability density function - the Gaussian in this case
    y = norm.pdf(x, mu, std)

    # Fit a normal distribution to the data:
    return mu, std, grades, x, y

def dual_plot(foo, grades, x, y, μ, σ, mu, std, annotation):
    
    # Plot the histogram.
    foo.hist(grades, bins=20, range=(0,max_grade), density=True, alpha=0.6, color='g')

    # Plot the PDF.
    foo.plot(x, y, 'k', linewidth=2)

    t1 = " Policy:  μ = %.2f,    σ = %.2f " % (μ, σ)
    t2 = " Output: mu = %.2f,  std = %.2f" % (mu, std)
    foo.set_title(annotation + t1 + t2)
```

```python
def dual_panel(pcnt_missed, assignments):
    missed = int(round(pcnt_missed/100.0 * assignments))
    fig = plt.figure()
    fig.suptitle(f' One Trial Comparison of Policy Effect for {missed} missed assignments out of {assignments}',
                 fontsize=16)
    mu, std, grades, x, y = gen_grades(μ, σ, assignments, missed, 'Orthodox')
    dual_plot(plt.subplot("221"), grades, x, y, μ, σ, mu, std, 'Orthodox')

    mu, std, grades, x, y = gen_grades(μ, σ, assignments, missed, 'Novel')
    dual_plot(plt.subplot("222"), grades, x, y, μ, σ, mu, std, 'Novel')
```
